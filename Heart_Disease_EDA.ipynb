```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

# Distribution plot for numeric columns

## Age Distribution

```python
sns.displot(x = df['Age'], color = 'blue',kde = True)
plt.title('Age Distribution')
plt.show()
```

## Distribution of RestingBP

```python
sns.displot(x = df['RestingBP'], kde = True, color = 'blue')
plt.title('Resting blood Pressure Distribution')
plt.show()
```

## Distribution of MaxHR

```python
sns.displot(x = df['MaxHR'], kde = True, color = 'blue')
plt.title('Distribution of Maxmium Heart Rate')
plt.show()
```

## Distribution of cholesterol

```python
sns.displot(x = df['Cholesterol'],kde = True, color = 'blue')
plt.title('Distribution of Cholestrol')
plt.show()
```

## Heart disease among male and felames

```python
sns.violinplot(x = 'HeartDisease' , y = 'Age', data = df)
plt.title('Age Distribution by Heart Disease Status')
plt.show()
```
